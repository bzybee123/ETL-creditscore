```python
import pandas as pd

file_path = r"C:\Users\shiva\Downloads\credit_customers.csv"
creditscore = pd.read_csv(file_path)

creditscore.head()

```

```python
creditscore.columns
```

```python
creditscore.drop(columns = ["checking_status", "other_parties", "own_telephone"], inplace = True, errors ='ignore')
```

```python
creditscore.columns
```

```python
creditscore.rename(columns = {"duration" : "loan_duration", "employment" : "employment_length", "housing" : "house_ownership"}, inplace = True, errors = "ignore" )
```

```python
creditscore.columns
```

```python
creditscore.duplicated()
```

```python
creditscore.isna().sum()
```

```python
creditscore["class"]
```

```python
creditscore["class"].unique()
```

```python
creditscore["class"] = creditscore["class"].apply(lambda x : 1 if x == "good"  else 0)
```

```python
creditscore.drop(columns = "Foreign_worker", inplace = True)
```

```python
creditscore.columns = [i.capitalize() for i in creditscore.columns]
```

```python
creditscore
```

```python
# Mapping Employment_length and Savings_status

employment_mapping = { "unemployed" : 0.0, '<1': 0.5, '1<=X<4': 2.5, '4<=X<7': 5.5, '>=7': 7.0} 
creditscore["Employment_length"] = creditscore["Employment_length"].map(employment_mapping)
print(creditscore)
```

```python
# Mapping 'Savings_status 

savings_mapping = {
    'no known savings': 0.0,
    '<100': 50.0,
    '100<=X<500': 300.0,
    '500<=X<1000': 750.0,
    '>=1000': 1000.0}
creditscore["Savings_status"] = creditscore["Savings_status"].map(savings_mapping)
```

```python
creditscore
```

```python
# List of columns to convert

creditscore.categorical_cols = ['Credit_history', 'Purpose', 'Personal_status', 'Property_magnitude',
                    'Other_payment_plans', 'House_ownership', 'Job']

# Convert the selected columns to 'category' data type
for col in categorical_cols:
    creditscore[col] = creditscore[col].astype('category')

```
